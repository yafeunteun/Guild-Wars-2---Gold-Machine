```python
def get_recipe_cost(recipe_id, qty=1):
    cost = 0 ## cost of the recipe
    gain = 0 ## return of the item corresponding to the recipe
    endpoint = "https://api.guildwars2.com/v2/recipes/{}".format(recipe_id)

    res = requests.get(endpoint, headers=headers)
    ingredients = res.json()['ingredients']
    output_item_id = res.json()['output_item_id']
    #pprint(res.json())
    
    for ingredient in ingredients:
        qty = ingredient['count']
        item_id = ingredient['item_id']
        endpoint = "http://api.guildwars2.com/v2/commerce/prices?ids={}".format(item_id)
        res = requests.get(endpoint, headers=headers)

        unit_price = res.json()[0]['sells']['unit_price']
        cost += unit_price*qty
    
    
    endpoint = "http://api.guildwars2.com/v2/commerce/prices?ids={}".format(output_item_id)
    res = requests.get(endpoint, headers=headers)
    #pprint(res.json())
    
    gain = res.json()[0]['buys']['unit_price']

    #print("Gain: {}".format(gain))
    #print("Costs: {}".format(cost))

    
    #print(ingredients) 
    return gain - cost
```
