```python
from collections import namedtuple
import requests
from bs4 import BeautifulSoup
from tqdm import tqdm_notebook
```

```python
recipes = list()
```

```python
BASE_URL = "https://db.dulfy.net"
```

```python
Recipe = namedtuple("Recipe", "name, rarity, url_recipe, url_item")
```

```python
recipes = list()
for i in tqdm_notebook(range(1, 20)):
    response = requests.get(f"{BASE_URL}/recipes/armorsmith/400-500/0/min_rating/desc/{i}")
    html_doc = response.text
    soup = BeautifulSoup(html_doc, 'html.parser')
    raw_recipes = soup.findAll("td", {"class": "name"})
    raw_items = soup.findAll("td", {"class": "icon"})
    if len(raw_recipes) > 0:
        for recipe, item in zip(raw_recipes, raw_items):
            name = recipe.text
            r = list(recipe.children)[0]
            i = list(item.children)[0]
            url_recipe = BASE_URL + r.attrs['href']
            rarity = r.attrs['class'][0].split('_')[0]
            item_id = i.attrs['src'].split('/')[-1].split('.')[0]
            url_item = f"{BASE_URL}/item/{item_id}"
            recipes.append(Recipe(name, rarity, url_recipe, url_item))
    else:
        break
```

```python
response = requests.get("https://db.dulfy.net/item/48752")
html_doc = response.text
soup = BeautifulSoup(html_doc, 'html.parser')
n_offers = soup.findAll("p", {"class": "offers"})
n_offers = int(n_offers[0].text.split('(')[-1].split(')')[0].split(" ")[0])
n_requests = soup.findAll("p", {"class": "offers"})
n_requests = int(n_requests[1].text.split('(')[-1].split(')')[0].split(" ")[0])
money_gold = soup.findAll("span", {"class": "money_gold"})
money_silver = soup.findAll("span", {"class": "money_silver"})
money_copper = soup.findAll("span", {"class": "money_copper"})
```

```python
gold = [10000 * int(x.text) for x in money_gold][:2]
silver = [100 * int(x.text) for x in money_silver][:2]
copper = [int(x.text) for x in money_copper][:2]
```

```python
sell_price = gold[0] + silver[0] + copper[0]
buy_price = gold[1] + silver[1] + copper[1]
print(f"Buy: {buy_price}, Sell: {sell_price}, Delta: {sell_price-buy_price}")
```
